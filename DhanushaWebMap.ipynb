```python
import geopandas as gpd
import pandas as pd
import plotly.express as px

# Raw URL of the GeoJSON file on GitHub
file_url = 'https://raw.githubusercontent.com/rkr1969/webmap/main/Landuse_Ward.geojson'

# Load the GeoJSON file directly from the URL
gdf = gpd.read_file(file_url)

# Group by 'Landuse' and sum the 'Area_hectare'
summary = gdf.groupby('Landuse')['Area_hectare'].sum().reset_index()

# Calculate the total area
total_area = summary['Area_hectare'].sum()

# Calculate the percentage of each land use category
summary['Percentage'] = (summary['Area_hectare'] / total_area) * 100

# Create an interactive bar chart using Plotly
fig = px.bar(
    summary,
    x='Landuse',
    y='Area_hectare',
    text='Percentage',  # Display percentage on top of bars
    title='Land Use Area Distribution in Division Forest Office Juridiction',
    labels={'Area_hectare': 'Area (Hectares)', 'Landuse': 'Land Use Category'},
    hover_data={'Percentage': ':.2f'}  # Show percentage with 2 decimal places on hover
)

# Customize layout
fig.update_traces(texttemplate='%{text:.2f}%', textposition='outside')  # Format percentage text
fig.update_layout(
    xaxis_title="Land Use Category",
    yaxis_title="Area (Hectares)",
    showlegend=False
)

# Show the interactive chart
fig.show()
```
